```python
mar2020_new=mar2020.drop(["start station name", "start station id", "start station latitude", "start station longitude", "end station id", 
                          "end station name","end station latitude","end station longitude", "bikeid" ],axis=1)
```

```python
mar2020_new
```

```python
mar2019_new=mar2019.drop(["start station name", "start station id", "start station latitude", "start station longitude", "end station id", 
                          "end station name","end station latitude","end station longitude", "bikeid" ],axis=1)
```

```python
mar2019_new
```

```python
MvsF_df1=pd.merge(feb2018_new,mar2020_new, how= "outer", on= ("tripduration","starttime","stoptime","usertype","birth year","gender"))
```
